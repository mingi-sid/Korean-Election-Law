```python
result_dict = {'district_name': '', 'district_info': '', 'candidates': [],
               'electors_num': '', 'voted_num': '', 'nullified_num': ''}
is_td_candidate_info = False
candidate_info = {}
is_resigned = False
is_td_district_stats = False
for tr in tmp.select('tr'):
    for td in tr.select('td'):
        if td.has_attr('colspan') and td['colspan'] == '4':
            # print text before <br>
            district_name = td.get_text(separator='\n').split('\n')[0]
            district_info = '\n'.join(td.get_text(separator='\n').split('\n')[1:])
            print(f'district_name: {district_name}')
            print(f'district_info: {district_info}')
            result_dict['district_name'] = district_name
            result_dict['district_info'] = district_info
            break
        if td.has_attr('rowspan') and td.text.isdigit():
            print('candidate_no: ', td.text)
            candidate_no = int(td.text)
            candidate_info['candidate_no'] = candidate_no
            is_td_candidate_info = True
            which_td_candidate_info = 'candidate_name'
            continue
        if is_td_candidate_info:
            print(f'{which_td_candidate_info}: ', td.text)
            if which_td_candidate_info == 'candidate_name':
                candidate_info[which_td_candidate_info] = td.text
                which_td_candidate_info = 'received_votes'
            elif which_td_candidate_info == 'received_votes':
                is_resigned = True if td.has_attr('colspan') else False # no votes
                candidate_info[which_td_candidate_info] = int(td.text.replace(',', '')) if not is_resigned else 0
                which_td_candidate_info = 'received_votes_rank'
                if is_resigned:
                    candidate_info[which_td_candidate_info] = None
                    which_td_candidate_info = 'party_name' # GOTO party_name
            elif which_td_candidate_info == 'received_votes_rank':
                candidate_info[which_td_candidate_info] = int(td.text[:-1])
                which_td_candidate_info = 'party_name'
            elif which_td_candidate_info == 'party_name':
                candidate_info[which_td_candidate_info] = td.find('a')['title']
                which_td_candidate_info = 'received_votes_ratio'
                if is_resigned:
                    candidate_info[which_td_candidate_info] = 0.0
                    which_td_candidate_info = 'is_elected'
                    candidate_info[which_td_candidate_info] = False
                    which_td_candidate_info = 'eol' # GOTO eol (end of line)
            elif which_td_candidate_info == 'received_votes_ratio':
                candidate_info[which_td_candidate_info] = float(td.text.split('%')[0])
                which_td_candidate_info = 'is_elected'
            elif which_td_candidate_info == 'is_elected':
                candidate_info[which_td_candidate_info] = True if ('당선' in td.text) else False
                which_td_candidate_info = 'eol'
            if which_td_candidate_info == 'eol':
                candidate_info['is_resigned'] = is_resigned
                is_td_candidate_info = False
                is_resigned = False
                result_dict['candidates'].append(candidate_info)
                candidate_info = {}
            continue
        if td.has_attr('rowspan') and '계' in td.text:
            is_td_district_stats = True
            which_td_district_info = 'electors_num'
            continue
        if is_td_district_stats:
            print(f'{which_td_district_info}: ', td.text)
            if '선거인 수' in td.text or \
                '투표 수' in td.text or \
                '무효표 수' in td.text or \
                '투표율' in td.text:
                continue
            if which_td_district_info == 'electors_num':
                result_dict[which_td_district_info] = int(td.text.replace(',', ''))
                which_td_district_info = 'voted_num'
            elif which_td_district_info == 'voted_num':
                result_dict[which_td_district_info] = int(td.text.replace(',', ''))
                which_td_district_info = 'nullified_num'
            elif which_td_district_info == 'nullified_num':
                result_dict[which_td_district_info] = int(td.text.replace(',', ''))
                is_td_district_stats = False
            continue

```

```python
def get_election_result_dict_by_district(table) -> dict:
    result_dict = {'district_name': '', 'district_info': '', 'candidates': [],
               'electors_num': '', 'voted_num': '', 'nullified_num': ''}
    is_td_candidate_info = False
    candidate_info = {}
    is_resigned = False
    is_td_district_stats = False
    for tr in table.select('tr'):
        for td in tr.select('td'):
            if td.has_attr('colspan') and td['colspan'] == '4':
                # print text before <br>
                district_name = td.get_text(separator='\n').split('\n')[0]
                district_info = '\n'.join(td.get_text(separator='\n').split('\n')[1:])
                print(f'district_name: {district_name}')
                #print(f'district_info: {district_info}')
                result_dict['district_name'] = district_name
                result_dict['district_info'] = district_info
                break
            if td.has_attr('rowspan') and td.text.isdigit():
                #print('candidate_no: ', td.text)
                candidate_no = int(td.text)
                candidate_info['candidate_no'] = candidate_no
                is_td_candidate_info = True
                which_td_candidate_info = 'candidate_name'
                continue
            if is_td_candidate_info:
                #print(f'{which_td_candidate_info}: ', td.text)
                if which_td_candidate_info == 'candidate_name':
                    candidate_info[which_td_candidate_info] = td.text
                    which_td_candidate_info = 'received_votes'
                elif which_td_candidate_info == 'received_votes':
                    text_without_comma = td.text.replace(',', '').replace('.', '')
                    is_resigned = False if text_without_comma.isdigit() else True # no votes
                    candidate_info[which_td_candidate_info] = int(text_without_comma) if not is_resigned else 0
                    which_td_candidate_info = 'received_votes_rank'
                    if is_resigned:
                        candidate_info[which_td_candidate_info] = None
                        which_td_candidate_info = 'party_name' # GOTO party_name
                        break
                elif which_td_candidate_info == 'received_votes_rank':
                    candidate_info[which_td_candidate_info] = int(td.text[:-1])
                    which_td_candidate_info = 'party_name'
                elif which_td_candidate_info == 'party_name':
                    party_name = td.find('a')['title']
                    if party_name in ['돈당', '국가혁명당', '혁명배당금당', '배당금당', '혁명당', '허경영당']:
                        party_name = '국가혁명배당금당'
                    candidate_info[which_td_candidate_info] = party_name
                    assert candidate_info[which_td_candidate_info] != ''

                    which_td_candidate_info = 'received_votes_ratio'
                    if is_resigned:
                        candidate_info[which_td_candidate_info] = 0.0
                        which_td_candidate_info = 'is_elected'
                        candidate_info[which_td_candidate_info] = False
                        which_td_candidate_info = 'eol' # GOTO eol (end of line)
                elif which_td_candidate_info == 'received_votes_ratio':
                    candidate_info[which_td_candidate_info] = float(td.text.split('%')[0])
                    which_td_candidate_info = 'is_elected'
                elif which_td_candidate_info == 'is_elected':
                    candidate_info[which_td_candidate_info] = True if ('당선' in td.text) else False
                    which_td_candidate_info = 'eol'
                if which_td_candidate_info == 'eol':
                    candidate_info['is_resigned'] = is_resigned
                    is_td_candidate_info = False
                    is_resigned = False
                    result_dict['candidates'].append(candidate_info)
                    candidate_info = {}
                continue
            if td.has_attr('rowspan') and '계' in td.text:
                is_td_district_stats = True
                which_td_district_info = 'electors_num'
                continue
            if is_td_district_stats:
                #print(f'{which_td_district_info}: ', td.text)
                if '선거인 수' in td.text or \
                    '유권자 수' in td.text or \
                    '투표 수' in td.text or \
                    '무효표' in td.text or \
                    '투표율' in td.text:
                    continue
                if which_td_district_info == 'electors_num':
                    result_dict[which_td_district_info] = int(td.text.replace(',', ''))
                    which_td_district_info = 'voted_num'
                elif which_td_district_info == 'voted_num':
                    result_dict[which_td_district_info] = int(td.text.replace(',', ''))
                    which_td_district_info = 'nullified_num'
                elif which_td_district_info == 'nullified_num':
                    result_dict[which_td_district_info] = int(td.text.replace(',', ''))
                    is_td_district_stats = False
                continue
    return result_dict
```

```python
full_data = {}
```

```python
from time import sleep
for i, wide_unit_name in enumerate(wide_unit):
    print(i, wide_unit_name)
    response = requests.get(base_url + '/' + requests.utils.quote(wide_unit_name))
    bs_html = bs(response.text, 'html.parser')
    election_result_list = []
    for table in bs_html.select('table'):
        if '선거인 수' in table.text or '유권자 수' in table.text:
            election_result = get_election_result_dict_by_district(table)
            election_result_list.append(election_result)
    full_data[wide_unit_name] = election_result_list
    sleep(5)
```

## 비례대표 선거 결과 받아오기 (광역 지자체별)

```python
base_url = 'https://namu.wiki/w/' + requests.utils.quote('제21대 국회의원 선거/비례대표')
response = requests.get(base_url)
bs_html = bs(response.text, 'html.parser')
```

```python
tmp = bs_html.select('table')[35]
proportional_representative_result_list = {}
for tr in tmp.find_all('tr'):
    party_info = {}
    for i, td in enumerate(tr.find_all('td')):
        if ' 비례대표' in td.text:
            wide_unit_name = td.text.split(' ')[0]
            print(wide_unit_name)
            proportional_representative_result_list[wide_unit_name] = {'parties': [], 'electors_num': None, 'voted_num': None, 'nullified_num': None}
            continue
        if '선거인수' in td.text:
            proportional_representative_result_list[wide_unit_name]['electors_num'] = int(td.text.split(':')[-1].strip().replace(',', ''))
            continue
        if '투표수' in td.text:
            proportional_representative_result_list[wide_unit_name]['voted_num'] = int(td.text.split(':')[-1].strip().replace(',', ''))
            continue
        if '무효표' in td.text:
            proportional_representative_result_list[wide_unit_name]['nullified_num'] = int(td.text.split(':')[-1].strip().replace(',', ''))
            continue
        if '기호' in td.text:
            break
        if i == 0 and td.text.isdigit():
            party_info['party_no'] = int(td.text)
            continue
        if i == 1:
            party_info['party_name'] = td.text
            continue
        if i == 2:
            party_info['received_votes'] = int(td.text.replace(',', ''))
            continue
        if i == 3:
            party_info['received_votes_ratio'] = float(td.text.split('%')[0])
            proportional_representative_result_list[wide_unit_name]['parties'].append(party_info)
            continue
```

```python
def get_representative_result_by_wide_unit(table) -> dict:
    proportional_representative_result_dict = {}
    for tr in table.find_all('tr'):
        party_info = {}
        for i, td in enumerate(tr.find_all('td')):
            if ' 비례대표' in td.text and '시군구별' not in td.text:
                wide_unit_name = td.text.split(' ')[0]
                print(wide_unit_name)
                proportional_representative_result_dict[wide_unit_name] = {'parties': [], 'electors_num': None, 'voted_num': None, 'nullified_num': None}
                continue
            if '선거인수' in td.text:
                proportional_representative_result_dict[wide_unit_name]['electors_num'] = int(td.text.split(':')[-1].strip().replace(',', ''))
                continue
            if '투표수' in td.text:
                proportional_representative_result_dict[wide_unit_name]['voted_num'] = int(td.text.split(':')[-1].strip().replace(',', ''))
                continue
            if '무효표' in td.text:
                proportional_representative_result_dict[wide_unit_name]['nullified_num'] = int(td.text.split(':')[-1].strip().replace(',', ''))
                continue
            if '기호' in td.text:
                break
            if i == 0 and td.text.isdigit():
                party_info['party_no'] = int(td.text)
                continue
            if i == 1:
                party_info['party_name'] = td.text
                continue
            if i == 2:
                party_info['received_votes'] = int(td.text.replace(',', ''))
                continue
            if i == 3:
                party_info['received_votes_ratio'] = float(td.text.split('%')[0])
                proportional_representative_result_dict[wide_unit_name]['parties'].append(party_info)
                continue
    return proportional_representative_result_dict
```

```python
full_proportional_representative_result = {}
for table in bs_html.select('table')[35:]:
    proportional_representative_result_dict = get_representative_result_by_wide_unit(table)
    full_proportional_representative_result.update(proportional_representative_result_dict)
```
